```python
import urllib.request
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from urllib.request import urlopen
import regex as re
import csv
```

```python
#read file
with open('urls.csv','r') as f:
    csv_raw_cont=f.read()

#split by line
split_csv=csv_raw_cont.split('\n')

#remove empty line
split_csv.remove('')

#specify separator
separator="/"
```

```python
for each in split_csv:

    #specify the row index
    url_row_index=0 #in our csv example file the url is the first row so we set 0

    #get the url
    url = each.split()[url_row_index]
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
    }
    r = requests.get(url, headers=headers)
    soup = bs(r.content, 'html5lib')
    
    field_names= ['Title', 'Paragraph']
    data = {}
    data["Title"] = soup.find("title").get_text(strip = True)
    Para = soup.find_all("p")
    Para = re.sub('<[^<]+?>', '',str(Para))
    Para = str(Para).encode('utf8').decode('ascii', 'ignore')
    data["Paragraph"] = Para
    
    list_all = []
    list_all.append(data)
    #print(list_all)
    print('--------uploaded',url)
    
    with open('Cleaned.csv', 'a') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames = field_names)
        writer.writeheader()
        writer.writerows(list_all)
```
